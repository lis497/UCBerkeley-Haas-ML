```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import root
import sympy as sy
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.tan(x)

# Mask out the points near the asymptotes to avoid vertical spikes
y[np.abs(np.cos(x)) < 0.05] = np.nan

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = \tan(x)$', color='purple')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)

# Vertical asymptotes
for k in range(-3, 4):
    plt.axvline(x=(np.pi/2 + k*np.pi), color='red', linestyle='--', alpha=0.6)

plt.title('Graph of $y = \\tan(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.grid(True)
plt.legend()
plt.xticks(
    ticks=[-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    labels=[r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', '0',
            r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
)
plt.show()
```

```python
x = np.linspace(-4, 4, 400)
f = x**3
f_prime = 3*x**2
f_double_prime = 6*x

# Plot the function and its derivatives
plt.figure(figsize=(10, 6))
plt.plot(x, f, label=r'$f(x) = x^3$', color='blue')
plt.plot(x, f_prime, label=r"$f'(x) = 3x^2$", color='green')
plt.plot(x, f_double_prime, label=r"$f''(x) = 6x$", color='red')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
#plt.axis("equal")
# Highlight point of inflection
plt.scatter(0, 0, color='purple', zorder=5)
plt.text(0.2, 5, 'Point of inflection', color='purple')

plt.title('Function and Derivatives of $f(x) = x^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.xlim(-4,4)
plt.show()
```

### Problem 1

#### Behavior of $f'$ and $f''$

Using the functions for `f`, `df`, and `ddf`, explore the value of the first and second derivative at points in domain $x \in [-2, 2]$.  Note that you already know the function is convex aka concave up.

- Is the first derivative on this domain increasing or decreasing?  Assign your solution as a string to `ans_1a`.

- Is the second derivative on this domain positive or negative?  Assign your solution as a string to `ans_1b`. 

```python
x = np.linspace(-2,2,50)
x[:5]
```

```python
def f(x):
    return x**2
```

```python
def df(x):
    h = 0.0000001
    return (f(x+ h) - f(x))/h
```

```python
def ddf(x):
    h = 0.0000001
    return (df(x + h) - df(x))/h
```

```python
plt.plot(x,f(x),label = 'f')
plt.plot(x,df(x), label = 'df')
plt.plot(x,ddf(x), label = 'ddf')
plt.grid()
plt.legend()
```

```python
ans_1a = 'increasing'
ans_1b = 'positive'
```

```python
print(f'First derivative at x = -1 and x = 1: {df(-1), df(1)}')
print(f'Second derivative at x = -1 and x = 1: {ddf(-1), ddf(1)}')
```

### Problem 2

#### Example with Data

Below, we load in the data from our earlier credit example and build the model

$$\hat{y} = \theta_0 X + \theta_1$$

and determine the mean squared error as

$$\frac{1}{n} \sum_{i = 1}^n (\hat{y} - y)^2$$

Use scikitlearn to determine the values for $\theta_0$ and $\theta_1$ that minimize the `mse` function.

```python
credit = pd.read_csv('codio_15_3_solution/data/Credit.csv', index_col=0)
X = credit['Rating'].values
y = credit['Balance'].values
```

```python
def mse(theta0, theta1):
  yhat = theta0*X + theta1
  error = y - yhat
  squared_error = np.sum(error**2)
  return 1/(2*X.shape[0]) * squared_error
```

```python
mse = np.vectorize(mse)
```

```python
theta1 = np.linspace(0, 6, 100)
theta0 = np.linspace(-400, -360, 100)
T1, T0 = np.meshgrid(theta1, theta0)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(T1, T0, mse(T1, T0), alpha = 0.2)
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_title(r'Mean Squared Error for $\theta_0$ and $\theta_1$')
ax.set_zticks([])
ax.set_yticks(np.arange(-400, -360, 10));
```

```python
lr = LinearRegression().fit(X.reshape(-1,1), y)
best_theta_0 = lr.coef_
best_theta_1 = lr.intercept_
```

```python
print(f'Equation with optimal parameters: y = {best_theta_0[0]: .2f}x + {best_theta_1: .2f}')
```
