### plt with df

### plot with plt

```python
plt.hist(x)
plt.title('Random Normal Hist')
```

### plot with ax

```python
# fig is the largest frame, ax is the subplots, ax.set_facecolor() uses CSS color
# y = 0.5 means the title position

fig, axs = plt.subplots(2,2, figsize=(17,8))
fig.suptitle('NBA Charts',fontsize = 20,fontweight = 'bold', y = -0.1)
fig.tight_layout(h_pad = 2)

axs[0,1].set_facecolor("#FFE4C4")
axs[1,0].set_facecolor("pink")
axs[1,1].set_facecolor("lightgray")
fig.set_facecolor("DarkSeagreen")
# bins means how many bars you want to seperate the data
axs[1,0].hist(x, edgecolor = 'orange', bins=30)
axs[1,0].set_title('Normal hist')

# the ax keyword inside df.plot() means you should put this plot inside which axs subplot. alpha means color transparency
df.plot(kind='hist', y = ['Number','Age','Weight','Salary'], ax = axs[1,1],edgecolor = 'orange', bins=20, alpha = .5)
#axs[1,1].set_ylim(0,60)
axs[1,1].set_title('Hist')
plt.show()
```
